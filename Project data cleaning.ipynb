```python
import numpy as np
import pandas as pd
data=pd.read_json('data.json')
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
dum= pd.get_dummies(data.commodity_name, dtype=int,drop_first=True)
dum
```

```python
df=pd.concat([data,dum],axis='columns')
df
```

```python
df1=df.drop(["commodity_name","reported_date"],axis='columns')
df1
```

```python
df1.shape
```

```python
df2=df1.drop(['min_price_rs_quintal','max_price_rs_quintal'],axis='columns')
df2
```

```python
X=df2.drop('modal_price_rs_quintal',axis='columns')
X
```

```python
y=df2['modal_price_rs_quintal']
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```
