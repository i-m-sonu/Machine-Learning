```python
import numpy as np
import pandas as pd
df = pd.read_excel("veg.xlsx")
df
```

```python
df.shape
```

```python
df.price.isnull().sum()
```

```python
df1=df.dropna(subset='price')
df1
```

```python
df2=df1.drop(['datesk','Date'],axis='columns')
df3=df2.reset_index(drop=True)
df3
```

```python
df3.isnull().sum()
```

```python
df3['Item Name'].unique()
```

```python
dum=pd.get_dummies(df3['Item Name'],dtype=int)
dum
```

```python
df4=pd.concat([df3,dum],axis='columns')
df4
```

```python
df5=df4.drop('Item Name',axis='columns')
df5.head()
```

```python
df5['price']=df5['price'].replace(0.0,pd.NA)
df5
```

```python
df6=df5.dropna(subset='price')
df6
```

```python
# df6.to_json('output.json', orient='records')
```

```python
df7=df6.reset_index(drop=True)
df7
```

```python
X=df7.drop('price', axis='columns')
X
```

```python
y=df7.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_train,y_train)
```

```python
# pip install joblib
import joblib
model='model.joblib'
joblib.dump(lr,model)
```
